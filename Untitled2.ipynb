```python
import requests
from bs4 import BeautifulSoup

urls = [
    "https://timesofindia.indiatimes.com/sports/cricket/india-tour-of-england/ind-vs-eng-4th-test-jasprit-bumrahs-inclusion-gives-team-ammunition-to-win-former-batting-coach-on-indias-chances-in-manchester/articleshow/122854644.cms",
    "https://timesofindia.indiatimes.com/technology/tech-news/worlds-largest-asset-manager-blackrock-to-employees-you-cannot-carry-your-phones-and-laptops-to-china-instead-/articleshow/122841635.cms",
    "https://timesofindia.indiatimes.com/sports/cricket/india-tour-of-england/ind-vs-eng-4th-test-pitch-report-what-to-expect-from-old-trafford-very-very-flat/articleshow/122851852.cms"
]

def get_article_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract main article paragraphs
    paragraphs = soup.find_all('p')
    article = ' '.join([p.get_text() for p in paragraphs])

    return article
```

```python
from gtts import gTTS

for i, url in enumerate(urls):
    print(f"Processing article {i+1}...")
    text = get_article_text(url)

    # Optional: shorten very long text
    text = text[:7000]  # gTTS has a limit

    # Convert to speech
    tts = gTTS(text=text, lang='en')
    filename = f"news_article_{i+1}.mp3"
    tts.save(filename)
    print(f"Saved: {filename}")
```
